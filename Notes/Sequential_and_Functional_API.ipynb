```python
import tensorflow as tf
from tensorflow.keras.utils import plot_model
```

## Sequential API

```python
# Method 1
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(2,)),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
plot_model(model, show_shapes=True)
```

```python
# Method 2
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(64, activation='relu', input_shape=(2,)))
model.add(tf.keras.layers.Dense(16, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```python
plot_model(model, show_shapes=True)
```

## Functional API

### Single Input and Single Output

Create Layers and Create Connections

```python
input = tf.keras.layers.Input( shape=(2,) )

hidden1 = tf.keras.layers.Dense(64, activation='relu')(input)
hidden2 = tf.keras.layers.Dense(16, activation='relu')(hidden1)

output = tf.keras.layers.Dense(1, activation='sigmoid')(hidden2)
```

Define Model

```python
model = tf.keras.Model( inputs=[input],
                        outputs = [output]  )
```

```python
plot_model(model, show_shapes=True)
```

### Multiple Inputs and Single Output

Create Layers and Create Connections

```python
input1 = tf.keras.layers.Input( shape=(2,) )
norm1 = tf.keras.layers.Normalization()(input1)
hidden1 = tf.keras.layers.Dense(16, activation='relu')(norm1)

input2 = tf.keras.layers.Input( shape=(5,) )
hidden2 = tf.keras.layers.Dense(32, activation='relu')(input2)

concat = tf.keras.layers.concatenate([hidden1, hidden2])

output = tf.keras.layers.Dense(1, activation='sigmoid')(concat)
```

Define Model

```python
model = tf.keras.Model(
    inputs = [input1, input2],
    outputs = [output]
)
```

```python
plot_model(model, show_shapes=True)  # Plot Model
```

### Single Input Multiple Outputs

```python
input = tf.keras.layers.Input( shape=(2,) )

hidden1 = tf.keras.layers.Dense(64, activation='relu')(input)
hidden2 = tf.keras.layers.Dense(32, activation='relu')(hidden1)

output1 = tf.keras.layers.Dense(1, activation='sigmoid')(hidden2)
output2 = tf.keras.layers.Dense(2, activation='softmax')(hidden2)
```

```python
model = tf.keras.Model(
    inputs = [input],
    outputs = [output1, output2]
)
```

```python
plot_model(model, show_shapes=True)  # Plot Model
```

### Multiple Input Multiple Output

```python
input1 = tf.keras.layers.Input( shape=(2,) )
norm1 = tf.keras.layers.Normalization()(input1)
hidden1 = tf.keras.layers.Dense(16, activation='relu')(norm1)

input2 = tf.keras.layers.Input( shape=(5,) )
hidden2 = tf.keras.layers.Dense(32, activation='relu')(input2)

concat = tf.keras.layers.concatenate([hidden1, hidden2])

output1 = tf.keras.layers.Dense(1, activation='sigmoid')(concat)
output2 = tf.keras.layers.Dense(3, activation='softmax')(concat)
```

```python
model = tf.keras.Model(
    inputs = [input1, input2],
    outputs = [output1, output2]
)
```

```python
plot_model(model, show_shapes=True)  # Plot Model
```
